```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
california = pd.read_csv("California_EUA.csv", encoding = "unicode_escape")
girona = pd.read_csv("Girona_España.csv", encoding = "unicode_escape")
mexico = pd.read_csv("DF_Mexico.csv") 
```

# Tratamiento de valores nulos

```python
california.describe()
```

```python
# Comprobamos los valores nulos de nuestro dataframe
pd.set_option('display.max_rows', None)
california.isna().sum()
```

### Variables categóricas 

```python
pd.reset_option('display.max_rows')
# Eliminamos esta columna debido a que todos los valores son nulos 
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
california = california.drop(['neighbourhood_group_cleansed','bathrooms', "calendar_updated"], axis=1)
california
```

```python
# Transfromaciones de variables 
california['host_acceptance_rate'] = california['host_acceptance_rate'].str.rstrip('%').astype('float') / 100.0
california['host_response_rate'] = california['host_response_rate'].str.rstrip('%').astype('float') / 100.0
california['price'] = california['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Se sustituyen los valores de la columna descripción por el string desconocido
california["description"] = california["description"].fillna("Desconocido")
california["neighborhood_overview"] = california["neighborhood_overview"].fillna("Desconocido")
california["host_location"] = california["host_location"].fillna("Desconocido")
# Hacemos el mismo proceso de una manera más compacta para las siguientes columnas 
california.loc[:, ["host_about", "host_response_time", "host_is_superhost"]] = california[["host_about", "host_response_time", "host_is_superhost"]].fillna("Desconocido")
# Realizamos una transformación para las variables: 
# Hacemos el mismo proceso de una manera más compacta para las siguientes col
california.loc[:, ["host_neighbourhood", "neighbourhood", "bathrooms_text", "license"]] = california[["host_neighbourhood", "neighbourhood", "bathrooms_text", "license"]].fillna("Desconocido")
# # Para la columna first review 
california.loc[:, ["first_review", "last_review"]] = california[["first_review", "last_review"]].fillna("00/00/00")
```

```python
# Comprobamos la sustitución 
pd.set_option('display.max_rows', None)
california.isna().sum()
```

### Variables cuantitativas

```python
# Realizamos el procesamiento para las variables numéricas de nuestro df
# Debido a que no conocemos estos datos los remplazaremos por la moda
california["bedrooms"] = california["bedrooms"].fillna(california["bedrooms"].mode()[0])
# Se realiza el mismo procedimiento con beds 
california["beds"] = california["beds"].fillna(california["beds"].mode()[0])

# Para las variables minimum_minimum_nights, maximum_minimum_nights, minimum_maximum_nights, maximum_maximum_nights, minimum_nights_avg_ntm, maximum_nights_avg_ntm, sustituimos con el dato que más se repite
cols = ["minimum_minimum_nights", "maximum_minimum_nights", "minimum_maximum_nights", "maximum_maximum_nights", "minimum_nights_avg_ntm", "maximum_nights_avg_ntm"]
california[cols] = california[cols].fillna({col: california[col].mode()[0] for col in cols})

# Hacemos lo mismo para la siguientes columnas que incluyen puntuaciones de las habitaciones 
cols = ["review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value", "reviews_per_month"]
california[cols] = california[cols].fillna({col: california[col].mode()[0] for col in cols})      

# Realizamos la sustitución para las columnas host_acceptance_rate host_response_rate
california['host_response_rate'] = california['host_response_rate'].fillna(california['host_response_rate'].mean())
california['host_acceptance_rate'] = california['host_acceptance_rate'].fillna(california['host_acceptance_rate'].mean())

# Para bathrooms text, haremos la siguiente transformación:
california['bathrooms_text'] = california['bathrooms_text'].str.extract('(\d+)').astype(float)
california["bathrooms_text"] = california["bathrooms_text"].fillna(california["bathrooms_text"].mode()[0])
```

```python
# Comprobamos la sustitución 
california.isna().sum()
```

# Tratamiento de outliers 

```python
# Seleccionamos los datos numéricos de nuestro df
pd.reset_option('display.max_rows')
california_numeric = california.select_dtypes(include=['int', 'float'])
print(california_numeric.columns)

# Seleccionamos las columnas categóricas 
california_cat = california.select_dtypes(include=["object"])
california_cat.columns
```

## Hacemos el tratamiento de outliers

#### Para el tratamiento de outliers para nuestras variables cuantitativas utilizaremos la sustitución por la mediana. Esto para evitar complicaciones en variables en donde las cantidades sean enteras. Acommodates, bedrooms, beds, etc. 

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y = california_numeric

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)

#Encontramos Outliers del Dataframe
outliers_iqr= california_numeric[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr

#Obtenemos datos limpios del Dataframe
data3_iqr= california_numeric[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr

#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos

#Reemplazamos valores atípicos (nulos) del dataframe con "median"
#Realizamos una copia del dataframe
data4_iqr= data3_iqr.copy()
california_numeric = data4_iqr.fillna(round(data4_iqr.median(),1))
```

```python
california_clean = pd.concat([california_cat, california_numeric], axis=1)
california_clean
```

```python
pd.set_option('display.max_rows', None)
california_clean.isna().sum().sum()
```

```python
import matplotlib.pyplot as plt

california_clean.boxplot(column=["host_acceptance_rate"])
plt.show()
```
